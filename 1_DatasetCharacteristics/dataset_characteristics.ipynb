```python
# Display column names grouped by category
print("="*80)
print("COLUMN DESCRIPTIONS")
print("="*80)

print("\n1. IDENTIFICATION & DATE:")
id_cols = ['id', 'date']
print(f"   {id_cols}")
print("   - Unique identifier and date information")

print("\n2. TARGET VARIABLES:")
target_cols = ['Warengruppe', 'umsatz']
print(f"   {target_cols}")
print("   - Product group and sales (revenue)")

print("\n3. ORIGINAL WEATHER DATA:")
weather_original = ['Temperatur', 'Bewoelkung', 'Wettercode', 'Windgeschwindigkeit']
print(f"   {weather_original}")
print("   - Raw weather measurements")

print("\n4. TEMPORAL FEATURES:")
temporal_cols = ['day_of_the_week', 'month', 'is_weekend', 'days_to_weekend', 'season', 'season_numeric']
print(f"   {temporal_cols}")
print("   - Day, month, weekend, and seasonal information")

print("\n5. HOLIDAY & EVENT FEATURES:")
holiday_cols = ['KielerWoche', 'Is_Holiday', 'Day_Before_Holiday', 'Day_After_Holiday', 
                'Is_Vacation', 'Vacation_Type']
print(f"   {holiday_cols}")
print("   - Special events, holidays, and vacation periods")

print("\n6. ENGINEERED WEATHER FEATURES:")
weather_eng = ['bewoelkung_category', 'wettercode_category', 'temperature_category',
               'temp_change_1d', 'temp_trend_3d']
print(f"   {weather_eng}")
print("   - Categorical and derived weather features")

print("\n7. LAG & TIME SERIES FEATURES:")
lag_cols = ['umsatz_lag1', 'umsatz_lag7', 'umsatz_pct_change', 'umsatz_pct_change_lag1']
print(f"   {lag_cols}")
print("   - Previous sales values and percentage changes")

print("\n8. ROLLING STATISTICS:")
rolling_cols = ['umsatz_ma7', 'umsatz_ma14', 'umsatz_volatility_7d', 'umsatz_sum_7d']
print(f"   {rolling_cols}")
print("   - Moving averages, volatility, and rolling sums")

print("\n9. INTERACTION FEATURES:")
interaction_cols = [col for col in df.columns if 'interaction' in col or 'combined' in col]
print(f"   {interaction_cols}")
print("   - Combined features from multiple variables")

print("\n10. NUMERIC ENCODINGS:")
encoded_cols = ['is_weekend_int'] + [col for col in df.columns if col not in 
    id_cols + target_cols + weather_original + temporal_cols + holiday_cols + 
    weather_eng + lag_cols + rolling_cols + interaction_cols]
print(f"   {encoded_cols}")
print("   - Numeric versions of categorical features")

print("\n" + "="*80)
print(f"TOTAL COLUMNS: {len(df.columns)}")
print("="*80)
```

## Missing Value Analysis

```python
# Calculate missing values
missing_counts = df.isnull().sum()
missing_percentages = (df.isnull().sum() / len(df)) * 100

# Create a summary DataFrame
missing_summary = pd.DataFrame({
    'Column': missing_counts.index,
    'Missing_Count': missing_counts.values,
    'Missing_Percentage': missing_percentages.values
})

# Sort by missing count (descending) and filter to show only columns with missing values
missing_summary = missing_summary[missing_summary['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

print("="*80)
print("MISSING VALUES SUMMARY")
print("="*80)
print(f"\nTotal rows in dataset: {len(df)}")
print(f"Columns with missing values: {len(missing_summary)}")
print(f"\nDetails:\n")
print(missing_summary.to_string(index=False))
```
