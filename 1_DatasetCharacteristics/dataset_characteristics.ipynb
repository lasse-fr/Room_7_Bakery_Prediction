### Key Observations:

1. **Intentional Tail Missingness (Expected):**
   - `umsatz` and derived lag/rolling features are missing at the end of the dataset (after 2017-12-27). This is expected as the target variable has no values beyond the training period, creating NaN for forward-looking features. This tail represents the hold-out/test set.

2. **Sparse Event Features (Expected):**
   - `KielerWoche`: Missing most of the year, present only during the event week (once annually). Normal and acceptable.
   - `Vacation_Type`: Missing when no vacation is occurring. Normal and acceptable.

3. **Concern: Weather Code Missingness (Throughout Dataset):**
   - `Wettercode` and `wettercode_category` have NaN values scattered throughout the entire dataset, not just at the tail.
   - This is problematic because weather is likely a key predictor of sales.
   - **Action Required**: Investigate whether missingness clusters in specific date ranges or is random, and consider external data sources or imputation strategies.


```python
# Check if Wettercode missingness clusters in time
print("="*80)
print("WETTERCODE MISSINGNESS: TEMPORAL PATTERN ANALYSIS")
print("="*80)

# Convert date to datetime if not already
df['date'] = pd.to_datetime(df['date'])

# Add a flag for missing Wettercode
df['wettercode_missing'] = df['Wettercode'].isnull()

# Group by year and month to see missingness pattern
missing_by_month = df.groupby(df['date'].dt.to_period('M'))['wettercode_missing'].agg(['sum', 'count'])
missing_by_month['missing_percentage'] = (missing_by_month['sum'] / missing_by_month['count'] * 100).round(2)
missing_by_month = missing_by_month.rename(columns={'sum': 'missing_count', 'count': 'total_records'})

print("\nMissing Wettercode by Month (showing non-zero months only):")
print(missing_by_month[missing_by_month['missing_count'] > 0].to_string())

# Plot temporal distribution of missing values
fig, ax = plt.subplots(figsize=(14, 5))
missing_by_date = df.groupby('date')['wettercode_missing'].mean() * 100
missing_by_date.plot(ax=ax, color='steelblue', linewidth=1)
ax.set_title('Wettercode Missingness Over Time (%)', fontsize=14, fontweight='bold')
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Missing Percentage (%)', fontsize=12)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Summary statistics
total_missing = df['wettercode_missing'].sum()
print(f"\n{'='*80}")
print(f"Total Wettercode Missing: {total_missing} ({total_missing/len(df)*100:.2f}%)")
print(f"Date Range: {df['date'].min()} to {df['date'].max()}")
print(f"\nClustering Assessment:")
if missing_by_month[missing_by_month['missing_count'] > 0].shape[0] <= 3:
    print("â†’ Missingness is CLUSTERED in specific months/periods")
    print("â†’ Consider sourcing external weather data for those periods OR excluding them from training")
else:
    print("â†’ Missingness is SCATTERED throughout the dataset")
    print("â†’ Consider imputation (forward-fill, seasonal mode, or external data)")
print(f"{'='*80}")
```

### 3.2 Temporal Features Distribution

```python
# Temporal features analysis
print("="*80)
print("TEMPORAL FEATURES ANALYSIS")
print("="*80)

# Day of week analysis with confidence intervals
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Day of week
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_means = []
day_ci = []

for day in day_order:
    day_data = df[df['day_of_the_week'] == day]['umsatz'].dropna()
    mean = day_data.mean()
    sem = stats.sem(day_data)
    ci = sem * stats.t.ppf((1 + 0.95) / 2., len(day_data) - 1)
    day_means.append(mean)
    day_ci.append(ci)

axes[0, 0].bar(range(len(day_order)), day_means, yerr=day_ci, capsize=5, color='steelblue', alpha=0.7, edgecolor='black')
axes[0, 0].set_xticks(range(len(day_order)))
axes[0, 0].set_xticklabels(day_order, rotation=45, ha='right')
axes[0, 0].set_title('Average Sales by Day of Week\n(with 95% CI)', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('Average Sales (â‚¬)', fontsize=10)
axes[0, 0].grid(True, alpha=0.3, axis='y')

# Month analysis
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_means = []
month_ci = []

for month in range(1, 13):
    month_data = df[df['month'] == month]['umsatz'].dropna()
    mean = month_data.mean()
    sem = stats.sem(month_data)
    ci = sem * stats.t.ppf((1 + 0.95) / 2., len(month_data) - 1)
    month_means.append(mean)
    month_ci.append(ci)

axes[0, 1].bar(range(len(month_order)), month_means, yerr=month_ci, capsize=5, color='coral', alpha=0.7, edgecolor='black')
axes[0, 1].set_xticks(range(len(month_order)))
axes[0, 1].set_xticklabels(month_order, rotation=45, ha='right')
axes[0, 1].set_title('Average Sales by Month\n(with 95% CI)', fontsize=12, fontweight='bold')
axes[0, 1].set_ylabel('Average Sales (â‚¬)', fontsize=10)
axes[0, 1].grid(True, alpha=0.3, axis='y')

# Season analysis
season_order = ['Spring', 'Summer', 'Fall', 'Winter']
season_means = []
season_ci = []

for season in season_order:
    season_data = df[df['season'] == season]['umsatz'].dropna()
    mean = season_data.mean()
    sem = stats.sem(season_data)
    ci = sem * stats.t.ppf((1 + 0.95) / 2., len(season_data) - 1)
    season_means.append(mean)
    season_ci.append(ci)

axes[1, 0].bar(range(len(season_order)), season_means, yerr=season_ci, capsize=5, color='lightgreen', alpha=0.7, edgecolor='black')
axes[1, 0].set_xticks(range(len(season_order)))
axes[1, 0].set_xticklabels(season_order, rotation=45, ha='right')
axes[1, 0].set_title('Average Sales by Season\n(with 95% CI)', fontsize=12, fontweight='bold')
axes[1, 0].set_ylabel('Average Sales (â‚¬)', fontsize=10)
axes[1, 0].grid(True, alpha=0.3, axis='y')

# Weekend analysis
weekend_data = df[df['is_weekend'] == 1]['umsatz'].dropna()
weekday_data = df[df['is_weekend'] == 0]['umsatz'].dropna()

weekend_means = [weekday_data.mean(), weekend_data.mean()]
weekend_ci = []

for data in [weekday_data, weekend_data]:
    sem = stats.sem(data)
    ci = sem * stats.t.ppf((1 + 0.95) / 2., len(data) - 1)
    weekend_ci.append(ci)

axes[1, 1].bar(['Weekday', 'Weekend'], weekend_means, yerr=weekend_ci, capsize=5, color='mediumpurple', alpha=0.7, edgecolor='black')
axes[1, 1].set_title('Average Sales: Weekday vs Weekend\n(with 95% CI)', fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel('Average Sales (â‚¬)', fontsize=10)
axes[1, 1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\nDay of Week Statistics:")
for day, mean, ci in zip(day_order, day_means, day_ci):
    print(f"  {day:12} - Mean: â‚¬{mean:>8,.0f}  (Â±â‚¬{ci:>7,.0f})")

print("\nMonth Statistics:")
for month, mean, ci in zip(month_order, month_means, month_ci):
    print(f"  {month:3} - Mean: â‚¬{mean:>8,.0f}  (Â±â‚¬{ci:>7,.0f})")

print("\nSeason Statistics:")
for season, mean, ci in zip(season_order, season_means, season_ci):
    print(f"  {season:8} - Mean: â‚¬{mean:>8,.0f}  (Â±â‚¬{ci:>7,.0f})")

print("\nWeekend vs Weekday:")
print(f"  Weekday - Mean: â‚¬{weekend_means[0]:>8,.0f}  (Â±â‚¬{weekend_ci[0]:>7,.0f})")
print(f"  Weekend - Mean: â‚¬{weekend_means[1]:>8,.0f}  (Â±â‚¬{weekend_ci[1]:>7,.0f})")
```

### 3.4 Special Events & Holiday Features

```python
# Holiday and special events analysis
print("="*80)
print("SPECIAL EVENTS & HOLIDAY FEATURES")
print("="*80)

# Holiday impact with confidence intervals
holiday_features = ['Is_Holiday', 'Day_Before_Holiday', 'Day_After_Holiday', 'KielerWoche']
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for idx, feature in enumerate(holiday_features):
    ax = axes[idx // 2, idx % 2]
    
    if df[feature].isna().all():
        ax.text(0.5, 0.5, f'No data for {feature}', ha='center', va='center', transform=ax.transAxes)
        ax.set_title(f'{feature} Distribution', fontsize=12, fontweight='bold')
        continue
    
    # For sparse event features (mostly NaN), treat NaN as 0 (event not happening)
    if feature == 'KielerWoche':
        df_temp = df.copy()
        df_temp[feature] = df_temp[feature].fillna(0)  # Treat NaN as 0 (no event)
        categories = [0, 1]
    elif feature in ['Is_Holiday', 'Day_Before_Holiday', 'Day_After_Holiday']:
        categories = [0, 1]
        df_temp = df
        # Filter to only include categories that actually exist in data
        categories = [c for c in categories if c in df[feature].dropna().unique()]
    else:
        df_temp = df
        categories = sorted(df[feature].dropna().unique())
    
    means = []
    cis = []
    
    for cat in categories:
        cat_data = df_temp[df_temp[feature] == cat]['umsatz'].dropna()
        if len(cat_data) > 0:
            mean = cat_data.mean()
            sem = stats.sem(cat_data)
            ci = sem * stats.t.ppf((1 + 0.95) / 2., len(cat_data) - 1)
            means.append(mean)
            cis.append(ci)
    
    # Create category labels
    if feature in ['Is_Holiday', 'Day_Before_Holiday', 'Day_After_Holiday', 'KielerWoche']:
        cat_labels = ['No', 'Yes'][:len(categories)]
    else:
        cat_labels = [str(c) for c in categories]
    
    ax.bar(range(len(means)), means, yerr=cis, capsize=5, color='mediumseagreen', alpha=0.7, edgecolor='black')
    ax.set_xticks(range(len(means)))
    ax.set_xticklabels(cat_labels, rotation=0)
    ax.set_title(f'{feature} Impact on Sales (with 95% CI)', fontsize=12, fontweight='bold')
    ax.set_ylabel('Average Sales (â‚¬)', fontsize=10)
    ax.grid(True, alpha=0.3, axis='y')
    
    # Print statistics
    print(f"\n{feature}:")
    for cat, mean, ci in zip(categories, means, cis):
        count = len(df_temp[df_temp[feature] == cat])
        cat_label = 'No' if cat == 0 and feature in ['Is_Holiday', 'Day_Before_Holiday', 'Day_After_Holiday', 'KielerWoche'] else ('Yes' if cat == 1 and feature in ['Is_Holiday', 'Day_Before_Holiday', 'Day_After_Holiday', 'KielerWoche'] else str(cat))
        print(f"  {feature}={cat_label}: Mean=â‚¬{mean:>8,.0f} (Â±â‚¬{ci:>7,.0f}), Count={count:>5}")

plt.tight_layout()
plt.show()

# Vacation analysis
print(f"\n{'='*80}")
print("VACATION ANALYSIS")
print(f"{'='*80}")

vacation_data = df[df['Is_Vacation'] == 1]['umsatz'].dropna()
non_vacation_data = df[df['Is_Vacation'] == 0]['umsatz'].dropna()

vac_mean = vacation_data.mean()
vac_ci = stats.sem(vacation_data) * stats.t.ppf((1 + 0.95) / 2., len(vacation_data) - 1)
non_vac_mean = non_vacation_data.mean()
non_vac_ci = stats.sem(non_vacation_data) * stats.t.ppf((1 + 0.95) / 2., len(non_vacation_data) - 1)

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(['Non-Vacation', 'Vacation'], [non_vac_mean, vac_mean], 
       yerr=[non_vac_ci, vac_ci], capsize=5, color='skyblue', alpha=0.7, edgecolor='black')
ax.set_title('Average Sales: Vacation vs Non-Vacation\n(with 95% CI)', fontsize=12, fontweight='bold')
ax.set_ylabel('Average Sales (â‚¬)', fontsize=10)
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print(f"Non-Vacation Days: Mean=â‚¬{non_vac_mean:>8,.0f} (Â±â‚¬{non_vac_ci:>7,.0f}), Count={len(non_vacation_data):>5}")
print(f"Vacation Days:     Mean=â‚¬{vac_mean:>8,.0f} (Â±â‚¬{vac_ci:>7,.0f}), Count={len(vacation_data):>5}")

# Vacation type breakdown
if df['Vacation_Type'].notna().any():
    print(f"\n{'='*80}")
    print("VACATION TYPE BREAKDOWN")
    print(f"{'='*80}")
    
    for vac_type in df['Vacation_Type'].dropna().unique():
        vac_type_data = df[df['Vacation_Type'] == vac_type]['umsatz'].dropna()
        mean = vac_type_data.mean()
        print(f"  {vac_type}: Mean=â‚¬{mean:>8,.0f}, Count={len(vac_type_data):>5}")
```

### 4.2 Multicollinearity Check

```python
# Check for highly correlated feature pairs (multicollinearity)
print("="*80)
print("MULTICOLLINEARITY CHECK")
print("="*80)
print("\nHighly Correlated Feature Pairs (|correlation| > 0.8):")
print("="*80)

# Get all correlations excluding target variable
features_without_target = [col for col in numeric_cols if col != 'umsatz']
corr_matrix_features = df[features_without_target].corr()

# Find highly correlated pairs
high_corr_pairs = []
for i in range(len(corr_matrix_features.columns)):
    for j in range(i+1, len(corr_matrix_features.columns)):
        corr_value = corr_matrix_features.iloc[i, j]
        if abs(corr_value) > 0.8:
            high_corr_pairs.append((
                corr_matrix_features.columns[i],
                corr_matrix_features.columns[j],
                corr_value
            ))

# Sort by absolute correlation value
high_corr_pairs.sort(key=lambda x: abs(x[2]), reverse=True)

if high_corr_pairs:
    for i, (feat1, feat2, corr) in enumerate(high_corr_pairs, 1):
        print(f"{i:2d}. {feat1:30s} â†” {feat2:30s} : {corr:>7.4f}")
    
    print(f"\nâš ï¸  Found {len(high_corr_pairs)} highly correlated feature pairs!")
    print("\nðŸ’¡ Recommendation:")
    print("   Consider removing one feature from each highly correlated pair during modeling")
    print("   to avoid multicollinearity issues.")
else:
    print("âœ… No severe multicollinearity detected (all |correlations| < 0.8)")
    print("   Dataset is suitable for regression modeling without feature removal.")

print("="*80)
```
